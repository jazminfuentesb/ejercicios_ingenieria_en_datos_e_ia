```python
# Programa que calcula el promedio de 5 calificaciones

# Inicializamos la variable donde se almacenará la suma total
suma = 0

# Ciclo repetitivo para solicitar 5 calificaciones
for i in range(1, 6):
    calificacion = float(input(f"Ingrese la calificación {i}: "))
    suma += calificacion  # Se acumulan las calificaciones

# Cálculo del promedio
promedio = suma / 5

# Mostrar resultado
print(f"El promedio final es {promedio}")
```

```python
# Programa que muestra la tabla de multiplicar de un número

# Solicitar número entero al usuario
numero = int(input("Ingrese un número entero: "))

# Ciclo repetitivo para generar la tabla del 1 al 10
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} × {i} = {resultado}")
```

```python
# Programa que cuenta cuántos números pares se ingresan

# Solicitar al usuario la cantidad de números que desea ingresar
cantidad = int(input("¿Cuántos números desea ingresar?: "))

# Inicializar contador de números pares
contador_pares = 0

# Bucle repetitivo según la cantidad indicada
for i in range(1, cantidad + 1):
    numero = int(input(f"Ingrese el número {i}: "))

    # Verificar si el número es par
    if numero % 2 == 0:
        contador_pares += 1  # Se incrementa el contador si es par

# Mostrar resultado
print(f"El total de números pares es: {contador_pares}")
```

```python
# Programa que determina la tarifa de transporte según la edad

# Solicitar edad al usuario
edad = int(input("Ingrese la edad: "))

# Validar que la edad no sea negativa
if edad < 0:
    print("Error: la edad no puede ser negativa.")
else:
    # Estructura selectiva encadenada
    if edad < 6:
        print("El boleto es gratis.")
    elif edad <= 17:
        print("El costo del boleto es $10.")
    elif edad <= 64:
        print("El costo del boleto es $20.")
    else:
        print("El costo del boleto es $8.")
```

```python
# Programa que suma únicamente los números positivos
# El programa termina cuando el usuario ingresa 0

# Inicializar la variable acumuladora
suma = 0

# Solicitar el primer número
numero = int(input("Ingrese un número entero (0 para terminar): "))

# Estructura repetitiva con condición de salida
while numero != 0:

    # Verificar si el número es positivo
    if numero > 0:
        suma += numero  # Se acumula solo si es positivo

    # Solicitar el siguiente número
    numero = int(input("Ingrese otro número (0 para terminar): "))

# Mostrar resultado final
print(f"La suma total de los números positivos es: {suma}")
```
