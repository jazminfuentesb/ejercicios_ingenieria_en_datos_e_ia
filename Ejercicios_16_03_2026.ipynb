```python
# Resolución mediante el método simplex
```

```python
#Bibliotecas
import numpy as np
```

```python
#Función simplex dada por el profesor

def pivotar(A, b, c, base):
    m, n = A.shape
    while True:
        # Calcula los costos reducidos.
        cb = c[base]
        z = cb.dot(A)
        costos_reducidos = c - z
        # Encuentra variable entrante (mínimo costo reducido).
        q = np.argmin(costos_reducidos)
        if costos_reducidos[q] >= 0:
            # Solución óptima encontrada.
            break
        # Encuentra variable saliente.
        direcciones = np.array([b[i] / A[i, q] if A[i, q] > 0 else np.inf for i in range(m)])
        p = np.argmin(direcciones)
        if direcciones[p] == np.inf:
            print("Problema no acotado.")
            return None
        # Pivotar.
        base[p] = q
        t = A[p, q]
        A[p, :] /= t
        b[p] /= t
        for i in range(m):
            if i != p:
                factor = A[i, q]
                A[i, :] -= factor * A[p, :]
                b[i] -= factor * b[p]
    # Solución óptima.
    x = np.zeros(n)
    x[base] = b
    return x

def simplex(c, A, b):
    """
    Resuelve el problema de maximización c^T x sujeto a Ax <= b, x >= 0.

    Argumentos:
    c -- Vector de coeficientes de la función objetivo.
    A -- Matriz de restricciones.
    b -- Vector del lado derecho de las restricciones.

    Retorna:
    x -- Solución óptima.
    """
    m, n = A.shape
    # Añade variables de holgura para convertir restricciones Ax <= b en igualdades.
    A = np.hstack([A, np.eye(m)])
    c = np.hstack([c, np.zeros(m)])
    base = list(range(n, n + m))
    return pivotar(A, b.copy(), c, base)

# Ejemplo de uso:
c = np.array([2, 3])  # Coeficientes de la función objetivo.
A = np.array([[1, 2], [4, 0], [0, 4]])  # Matriz de restricciones.
b = np.array([8, 16, 16])  # Lado derecho de las restricciones.

solucion = simplex(c, A, b)
print("Solución óptima:", solucion)
```

## Ejercicio 1

```python
# ============================================
# EJERCICIO 1: Maximización de Beneficios
# ============================================
# Contexto:
# Una empresa produce dos productos, A y B.
# Cada unidad de A genera $2 y cada unidad de B genera $3.
#
# Restricciones:
# 1. Máquina: máximo 480 minutos
#    - A requiere 1 min/unidad
#    - B requiere 2 min/unidad
#
# 2. Trabajo manual: máximo 960 minutos
#    - A requiere 4 min/unidad
#    - B no requiere trabajo manual
#
# 3. Empaque: máximo 960 minutos
#    - A no requiere empaque
#    - B requiere 4 min/unidad
#
# Objetivo:
# Maximizar Z = 2x1 + 3x2
#
# Restricciones matemáticas:
# x1 + 2x2 <= 8
# 4x1 <= 16
# 4x2 <= 16
# x1, x2 >= 0
```

```python
# Función objetivo (negativa porque el código minimiza)
c = np.array([-2, -3])

# Restricciones
A = np.array([
    [1, 2],
    [4, 0],
    [0, 4]
])

b = np.array([8, 16, 16])

# Resolver con simplex
solucion = simplex(c, A, b)

# Extraer variables reales
x1 = solucion[0]
x2 = solucion[1]

# Calcular valor óptimo real
Z = 2*x1 + 3*x2

print("===== RESULTADOS =====")
print(f"Solución óptima: x1 = {x1:.2f}, x2 = {x2:.2f}")
print(f"Valor óptimo Z = {Z:.2f}")

print("\nInterpretación:")
print(f"La empresa debe producir {x1:.0f} unidades del producto A y {x2:.0f} unidades del producto B para maximizar sus beneficios.")

print("\nVerificación:")
print("Restricción 1 (máquina):", x1 + 2*x2, "<= 8")
print("Restricción 2 (trabajo manual):", 4*x1, "<= 16")
print("Restricción 3 (empaque):", 4*x2, "<= 16")
```

## Ejercicio 2

```python
# ============================================
# EJERCICIO 2: Problema de Dieta Óptima
# ============================================
# Contexto:
# Una empresa produce dos productos, P1 y P2.
#
# Beneficios:
# P1 genera $4
# P2 genera $5
#
# Restricciones:
# 1. Mano de obra: máximo 120 horas
# 2. Materia prima: máximo 160 kg
#
# Modelo:
# Maximizar Z = 4x1 + 5x2
#
# Restricciones:
# x1 + 2x2 <= 120
# 2x1 + x2 <= 160
# x1, x2 >= 0
```

```python
# Función objetivo (negativa porque el código minimiza)
c = np.array([-4, -5])

# Restricciones
A = np.array([
    [1, 2],
    [2, 1]
])

b = np.array([120, 160])

# Resolver con simplex
solucion = simplex(c, A, b)

# Extraer variables reales
x1 = solucion[0]
x2 = solucion[1]

# Calcular valor óptimo real
Z = 4*x1 + 5*x2

print("===== RESULTADOS =====")
print(f"Solución óptima: x1 = {x1:.2f}, x2 = {x2:.2f}")
print(f"Valor óptimo Z = {Z:.2f}")

print("\nInterpretación:")
print(f"Se deben producir {x1:.2f} unidades de P1 y {x2:.2f} unidades de P2 para maximizar el beneficio.")

print("\nVerificación:")
print("Restricción 1:", x1 + 2*x2, "<= 120")
print("Restricción 2:", 2*x1 + x2, "<= 160")
```

## Ejercicio 3

```python
# ============================================
# EJERCICIO 3: Asignación de Recursos
# ============================================
# Contexto:
# Se asignan recursos a dos proyectos (P1 y P2).
#
# Beneficios:
# Cada unidad asignada a P1 genera $2
# Cada unidad asignada a P2 genera $3
#
# Restricciones:
# Para P1:
# x1 <= 8
# Para P2:
# x2 <= 16
#
# Modelo:
# Maximizar Z = 2x1 + 3x2
#
# Restricciones:
# x1 <= 8
# x2 <= 16
# x1, x2 >= 0
```

```python
# Función objetivo (negativa porque el código minimiza)
c = np.array([-2, -3])

# Restricciones
A = np.array([
    [1, 0],   # x1 <= 8
    [0, 1],   # x2 <= 16
    [1, 0],   # x1 <= 16 (redundante)
    [0, 1]    # x2 <= 16 (redundante)
])

b = np.array([8, 16, 16, 16])

# Resolver con simplex
solucion = simplex(c, A, b)

# Extraer variables reales
x1 = solucion[0]
x2 = solucion[1]

# Calcular valor óptimo real
Z = 2*x1 + 3*x2

print("===== RESULTADOS =====")
print(f"Solución óptima: x1 = {x1:.2f}, x2 = {x2:.2f}")
print(f"Valor óptimo Z = {Z:.2f}")

print("\nInterpretación:")
print(f"Se deben asignar {x1:.0f} unidades al proyecto P1 y {x2:.0f} unidades al proyecto P2 para maximizar el beneficio.")

print("\nVerificación:")
print("Restricción 1:", x1, "<= 8")
print("Restricción 2:", x2, "<= 16")
print("Restricción 3:", x1, "<= 16")
print("Restricción 4:", x2, "<= 16")
```
